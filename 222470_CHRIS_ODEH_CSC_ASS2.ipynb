Importing Needed libraries

```python
import pandas as pd
```

Reading Data

```python
path = "https://scipython.com/static/media/2/problems/P2.6/redwood-data.txt"
```

```python
data = pd.read_csv(path,sep="\t",header = None, names=["Tree name","Location","Diameter(m)","Height(m)"])
data
```

Dropping NAN values

```python
new_data = data.dropna()
new_data
```

Finding the tree with the largest diameter

```python
#This code prints out the entire row of the tree with the widest diameter and drops the height column
wide_diameter = new_data[new_data["Diameter(m)"]==new_data["Diameter(m)"].max()]
wide_diametern = wide_diameter.drop(columns = "Height(m)")
print(wide_diametern)
```

Finding the tallest tree

```python
#This code checks for the tallest tree and prints out the entire row and then drops the diameter column

large_tree = new_data[new_data["Height(m)"]==new_data["Height(m)"].max()]
large_treen = large_tree.drop(columns="Diameter(m)")
print(large_treen)
```
